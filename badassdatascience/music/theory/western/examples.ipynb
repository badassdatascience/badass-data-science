## Load application-specific modules

```python
from badassdatascience.music.theory.Pitch import Pitch
import badassdatascience.music.theory.western.scales.scales as sc
from badassdatascience.music.theory.western.frequency.frequency_calculations import calculate_frequency
```

## Define a class to define pitches by MIDI number

```python
class PitchByMidiNumber(Pitch):

    def __init__(self, midi_number):
        self.midi_number = np.uint8(midi_number)
        
        self.octave = np.int8(np.floor(self.midi_number / 12.) - 1)
        self.frequency = calculate_frequency(self.midi_number)
        
        self.pitch_class_number = self.midi_number % 12
        self.pitch_class = sc.chromatic_scale_basic[self.pitch_class_number]
```

## See if it works!

```python
for i in [127, 120, 100, 69, 60, 59, 50, 34, 24, 14, 13, 12, 11, 1, 0]:
    p = PitchByMidiNumber(i)
    print(p.midi_number, p.octave, p.pitch_class)
```
